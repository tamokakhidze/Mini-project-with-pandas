```python
df = df.drop_duplicates()
df
```

```python
df = df.drop(columns = "Depth Error")
df
```

```python
bins = [float('-inf'), 3, float('inf')]
labels = ['Weak Earthquake', 'Strong Earthquake']
```

```python
df['Time'] = pd.to_datetime(df['Time'], format="%H:%M:%S", utc=True)
```

```python
df['Hour'] = df['Time'].dt.hour
```

```python
hourly_counts = df['Hour'].value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
plt.pie(hourly_counts, labels=hourly_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Earthquakes Distribution by Hour')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Longitude'], bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Earthquake Longitudes')
plt.xlabel('Longitude')
plt.ylabel('Frequency')
plt.show()
```

```python
df3 = df.head(50).sort_values(by='Magnitude')
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df3['Magnitude'], marker='o', linestyle='-', color='b')
plt.title('Line Graph of Earthquake Magnitudes')
plt.xlabel('Index')
plt.ylabel('Magnitude')
plt.show()
```

```python
status_counts = df['Status'].value_counts()
```

```python
plt.figure(figsize=(10, 6))
status_counts.plot(kind='bar', color='skyblue')
plt.title('Bar Plot of Earthquake Status')
plt.xlabel('Status')
plt.ylabel('Frequency')
plt.show()
```
